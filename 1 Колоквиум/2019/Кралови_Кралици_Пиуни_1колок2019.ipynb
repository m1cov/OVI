Две тројки, црна и бела, сочинети од крал, кралица и пиун треба да ја преминат реката.

Кралевите не смеат да останат без своите поданици на една страна од реката.
Кралиците сакаат да бидат сами кога ја преминуваат реката.
Пиуните и кралиците знаат да веслаат, кралевите не веслаат.
Чамецот може да издржи највеќе двајца.
Напишете алгоритам кој ќе го пронајде најмалиот број преминувања на реката, така што сите патници ќе преминат од левата на десната страна од реката.

```python
from collections import deque
from itertools import permutations
```

```python
permitted_states = [(0,),(0,3),(0,4),(0,5),(0,3,4),(0,3,5),(0,4,5),(0,3,4,5),(3,),(1,3),(2,3),(0,1,3),(0,2,3),(1,2,3),(0,1,2,3)] #mora po redosled da se indeksite
#permitted_states = [(0,),(0,3),(0,4),(0,5),(0,3,4),(0,3,5),(0,4,5),(0,3,4,5),(3,),(3,1),(3,2),(3,0,1),(3,0,2),(3,1,2),(3,0,1,2)]
```

```python
in_boat = [(1,),(4,),(2,),(5,),(2,5),(0,2),(3,5)] # (k1,q1,p1,k2,q2,p2,boat)
```

```python
available_moves = set()

moves = permutations([0, 0, 0, 0, 0, 0, 1], 7)
for i in moves:
    available_moves.add(i)

moves = permutations([0, 0, 0, 0, 0, 1, 1], 7)
for i in moves:
    available_moves.add(i)

moves = permutations([0, 0, 0, 0, 1, 1, 1], 7)
for i in moves:
    available_moves.add(i)

moves = permutations([0, 0, 0, 1, 1, 1, 1], 7)
for i in moves:
    available_moves.add(i)

moves = permutations([0, 0, 1, 1, 1, 1, 1], 7)
for i in moves:
    available_moves.add(i)

moves = permutations([0, 1, 1, 1, 1, 1, 1], 7)
for i in moves:
    available_moves.add(i)

available_moves.add((0, 0, 0, 0, 0, 0, 0))
available_moves.add((1, 1, 1, 1, 1, 1, 1))

available_moves = list(available_moves)
```

```python
def search_path(initial_state, goal_state):
    visited = {initial_state}
    queue = deque([[initial_state]])
    while queue:
        state_list = queue.popleft()
        state_to_expand = state_list[-1]
        for next_state in expand_state(state_to_expand):
            if next_state not in visited:
                if next_state == goal_state:
                    return state_list + [next_state]
                visited.add(next_state)
                queue.append(state_list + [next_state])
    return []
```

```python
def expand_state(state):
    new_states = []
    for new_state in available_moves:
        if is_valid(state, new_state):
            new_states.append(new_state)
    return new_states
```

```python
def is_valid(state, new_state):
    if state[-1] == new_state[-1]:
        return False
    else:
        helper = []
        for i in range(len(state)-1):
            if state[i] != state[-1] and state[i] != new_state[i]:
                return False
            else:
                if state[i] != new_state[i]:
                    helper.append(i)
            
        for passengers in in_boat:
            if len(passengers) == len(helper):
                for i in helper:
                    counter = False
                    if i not in passengers:
                        break
                    counter = True
                if counter:
                    if check_allowed_state(new_state):
                        return True
    return False 
```

```python
def check_allowed_state(new_state):
    left = [i for i in range(len(new_state)-1) if new_state[i] == 0]
    right = [i for i in range(len(new_state)-1) if new_state[i] == 1]
    if tuple(left) in permitted_states:
        return False
    elif tuple(right) in permitted_states:
        return False
    else:
        return True
```

```python
initial_state = (0,0,0,0,0,0,0)
goal_state = (1,1,1,1,1,1,1)
path = search_path(initial_state, goal_state)
```

```python
names = ['w_king','w_queen','w_pawn','b_king','b_queen','b_pawn','boat']
```
