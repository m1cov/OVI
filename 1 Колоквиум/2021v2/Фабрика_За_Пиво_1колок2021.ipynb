Разгледуваме фабрика за производство на пиво. За пакување на пивото во шишиња се користат рециклирани шишиња кои во фабриката пристигаат во гајби. Следат постапките:

Сите шишиња се ставаат врз платформата. Платформата собира 15х15 шишиња. За поедноставни пресметки за време на испитот, ќе сметаме дека платформата собира 7х7 шишиња. Софтвер за препознавање ги скенира сите шишиња од платформата, и во себе забележува по еден код за секое шише: (0) - Празно место, нема шише. (1) - Скршено шише. (2) - Шише во добра состојба, од нашата пивара. (3) - Шише од друга пивара. Роботот ги избира соодветните шишиња (со код 2) и ги префрла на подвижната лента за понатаму. Роботот наеднаш може да земе 10 шишиња, бидејќи има 10 дланки (5 х 2). За поедноставни пресметки за време на испитот, ќе сметаме дека роботот има 6 дланки (3 х 2), па роботот може наеднаш да земе најмногу 6 шишиња од платформата. Роботот може да ротира за 90 степени. Ваша задача е да напишете код со алгоритмите од вештачка интелигенција за чекорот 3. Целта е роботот во минимален број придвижувања да ги издвои соодветните шишиња (со код 2).

```python
from collections import deque
from copy import deepcopy
```

```python
g  = (
    (0, 3, 2, 0, 2, 2, 2),
    (2, 1, 2, 2, 2, 3, 2),
    (2, 2, 0, 2, 2, 2, 2),
    (0, 2, 2, 2, 2, 2, 0),
    (2, 2, 3, 2, 2, 2, 2),
    (2, 2, 1, 2, 2, 2, 2),
    (0, 3, 0, 2, 2, 2, 2)
)
```

```python
def search_path(initial_state,alg):
    visited = {initial_state}
    states_queue = deque([[initial_state]])
    while states_queue:
        states_list = states_queue.popleft()
        state_to_expand = states_list[-1]
        for next_state in expand_state1(state_to_expand):
            if next_state not in visited:
                if check_end(next_state):
                    return states_list + [next_state]
                visited.add(next_state)
                if alg == 'bfs':
                    states_queue.append(states_list + [next_state])
                if alg == 'dfs':
                    states_queue.appendleft(states_list + [next_state])
    return []
```

```python
def check_end(next_state):
    for i in range(7):
        if next_state[i].count(2) != 0:
            return False
    return True
```

```python
def find_next_two(state):
    for i, row in enumerate(state):
        for j, value in enumerate(row):
            if value == 2:
                return i, j
    return None
```

```python
def pick(x, y):
    neighbours = []
    check_1 = 1
    check_2 = 1
    check_3 = 1
    check_4 = 1
    neighbours_2x3_1 = [(x, y), (x, y+1), (x, y+2), (x+1, y), (x+1, y+1), (x+1, y+2)]
    neighbours_2x3_2 = [(x, y), (x, y-1), (x, y-2), (x+1, y), (x+1, y-1), (x+1, y-2)]
    neighbours_3x2_1 = [(x, y), (x+1, y), (x+2, y), (x, y+1), (x+1, y+1), (x+2, y+1)]
    neighbours_3x2_2 = [(x, y), (x+1, y), (x+2, y), (x, y-1), (x+1, y-1), (x+2, y-1)]
    for x, y in neighbours_2x3_1:
        if x>=7 or x<0 or y>=7 or y<0:
            check_1 = 0
    for x, y in neighbours_2x3_2:
        if x>=7 or x<0 or y>=7 or y<0:
            check_2 = 0
    for x, y in neighbours_3x2_1:
        if x>=7 or x<0 or y>=7 or y<0:
            check_3 = 0
    for x, y in neighbours_3x2_2:
        if x>=7 or x<0 or y>=7 or y<0:
            check_4 = 0
    if check_1:
        neighbours.append(neighbours_2x3_1)
    if check_2:
        neighbours.append(neighbours_2x3_2)
    if check_3:
        neighbours.append(neighbours_3x2_1)
    if check_4:
        neighbours.append(neighbours_3x2_2)
    return neighbours
```

```python
def expand_state1(state):
    new_states = []
    state_list = list([list(x) for x in state])
    x, y = find_next_two(state_list)
    neighbours_list = pick(x, y)
    for i in range(len(neighbours_list)):
        helper = deepcopy(state_list)
        for x, y in neighbours_list[i]:
            if state_list[x][y] == 2:
                helper[x][y] = 0
        new_states.append(tuple([tuple(row) for row in helper]))
    return new_states
```

```python
%%time
path = search_path(g,'bfs')
for state in path:
    for row in state:
        print(row)
    print()
```

## Решение со A* пребарување

```python
import heapq
```

```python
def a_star_search(initial_state, heuristic_function, alpha=1): 
    queue = [((0,0), [initial_state])]
    heapq.heapify(queue)
    expanded = set()
    while queue:
        weight_tuple, state_list = heapq.heappop(queue)
        current_a_star_weight, current_path_weight = weight_tuple
        state_to_expand = state_list[-1]
        if state_to_expand not in expanded:
            if check_end(state_to_expand):
                return current_path_weight, state_list
            for next_state, new_path_weight in expand_state2(state_to_expand):
                if next_state not in expanded:
                    path_weight = current_path_weight + new_path_weight
                    heuristic = heuristic_function(next_state)
                    a_star_weight = path_weight + alpha*heuristic
                    heapq.heappush(queue, ((a_star_weight, path_weight), state_list + [next_state]))
            expanded.add(state_to_expand)
```

```python
def expand_state2(state):
    new_states = []
    state_list = list([list(x) for x in state])
    x, y = find_next_two(state_list)
    neighbours_list = pick(x, y)
    for i in range(len(neighbours_list)):
        helper = deepcopy(state_list)
        #counter = 0
        for x, y in neighbours_list[i]:
            if state_list[x][y] == 2:
                helper[x][y] = 0
                #counter += 1
        new_states.append((tuple([tuple(row) for row in helper]),1))
    return new_states
```

```python
def heuristic(next_state):
    total = 0
    for row in next_state:
        for value in row:
            if value == 2:
                total += 1
    return total/6
```
